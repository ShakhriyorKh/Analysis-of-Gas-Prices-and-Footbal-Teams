### Load Required Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Real Statistics


#### Line Graph



```python
gas = pd.read_csv('Data/gas_prices.csv')

plt.figure(figsize=(8, 5))
plt.title('Oil Prices Over 1990-2007', fontdict={'fontweight': 'bold', 'fontsize': 15})
plt.plot(gas['Year'], gas['USA'], 'b.-', label='USA')
plt.plot(gas['Year'], gas['Canada'], 'r.-', label='Canada')
plt.plot(gas['Year'], gas['South Korea'], 'g.-', label='South Korea')
plt.plot(gas['Year'], gas['Australia'], 'y.-', label='Australia')

plt.xticks(gas['Year'][::2])


plt.xlabel('Years')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('1.oil_price_figure.jpg', dpi=300)

plt.show()
```

### Load FIFA Data

```python
fifa = pd.read_csv('Data/fifa_data.csv')
```

```python
bins = [ 40, 50, 60, 70, 80, 90, 100]

plt.figure(figsize=(8, 5))

plt.title('Number of Football Players by Age Distribution', fontdict={'fontsize': 15, 'fontweight':'bold'})
plt.hist(fifa['Overall'], bins=bins)

plt.xticks(bins)

plt.xlabel('Age of players')
plt.ylabel('Number pf Players')

plt.savefig('2.number_byAge.jpg', dpi=300)

plt.show()
```

### Pie Chart

```python
left_foot = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right_foot = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']
colors = ['Yellow', 'Gray']
plt.pie([left_foot, right_foot], labels=labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players', fontdict={'fontweight': 'bold', 'fontsize':15})

plt.savefig('3.foot_preference.jpg', dpi=300)

plt.show()
```

```python
fifa['Weight'] = [int(x.strip('lbs')) if type(x) == str else x for x in fifa['Weight']]

plt.figure(figsize=(7, 5))
plt.title('Weight Distribution of Players', fontdict={'fontsize': 15, 'fontweight': 'bold'})

light = fifa.loc[fifa['Weight'] < 125].count()[0]
light_medium = fifa.loc[(fifa['Weight'] >= 125) & (fifa['Weight'] < 150)].count()[0]
medium = fifa.loc[(fifa['Weight'] >= 150) & (fifa['Weight'] < 175)].count()[0]
medium_heavy = fifa.loc[(fifa['Weight'] >= 175) & (fifa['Weight'] < 200)].count()[0]
heavy = fifa.loc[(fifa['Weight'] >= 200) ].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = [.4, .1, .0, .0, .4]

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)

plt.savefig('4.weight_distribution.jpg', dpi=300)

plt.show()
```

```python
barcelona = fifa.loc[fifa['Club'] == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa['Club'] == 'Real Madrid']['Overall']

plt.figure(figsize=(8, 5))

labels = ['FC Barcelona', 'Real Madrid']
boxes = plt.boxplot([barcelona, madrid], labels=labels, patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    box.set(color='#4286f4', linewidth=2, facecolor='#e0e0e0')
    
plt.title('Football Team Comparison', fontdict={'fontsize': 15, 'fontweight': 'bold'})

plt.savefig('5.team_comparison.jpg', dpi=300)

plt.show()
```
